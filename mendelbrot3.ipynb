```python
import numpy as np
import matplotlib.pyplot as plt

DEPTH = 1000
CICLE = 15

im = np.linspace(-1,1,DEPTH)
re = np.linspace(-1.5,0.5,DEPTH)
```

```python
for y in im:
    for x in re:
        z,c = 0, complex(x,y)
        counter = 0
        while counter<CICLE and np.abs(z)<2:
            z = z*z + c
            counter += 1
        if counter == CICLE:
            plt.plot(x,y,'o',markersize=0.1,color='black')

plt.show()
```
